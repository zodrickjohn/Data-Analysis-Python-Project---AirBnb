## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Import Raw Data

```python
df = pd.read_csv("airbnb.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

## Data Cleaning

```python
df.isnull().sum()
```

```python
# dropping null values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
# dropping duplicate values
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df["id"] = df["id"].astype(object)
df["host_id"] = df["host_id"].astype(object)
df.dtypes
```

## Univariate Analysis

```python
sns.boxplot(data=df, x= "price")
```

```python
# deleting outliers
df2 = df[df["price"] < 1500]
sns.boxplot(data=df2, x= "price")
```

### Price distribution

```python
sns.histplot(data=df2, x= "price",bins=100)
plt.title("Price distribution")
plt.ylabel("Frequency")
plt.show()
```

### Availability_365 distribution

```python
sns.histplot(data=df2, x= "availability_365")
plt.title("Availability_365 distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
df2.groupby("neighbourhood_group")["price"].mean()
```

## Feature Engineering

```python
df2.head()
```

```python
df2["price per bed"] = df2["price"]/df2["beds"]
```

```python
df2.head()
```

```python
df2.groupby("neighbourhood_group")["price per bed"].mean()
```

## Bivariate Analysis

```python
df2.columns
```

### Price ditribution over different types of Rooms in different cities

```python
sns.barplot(data=df2, x="neighbourhood_group", y="price", hue="room_type")
```

### Locality and Review Dependency

```python
plt.figure(figsize=(8,5))
plt.title("Locality and Review Dependency")
sns.scatterplot(data = df2, x = "number_of_reviews", y="price", hue = "neighbourhood_group")
```

### Pairplot

```python
sns.pairplot(data=df2, vars=["price","minimum_nights","number_of_reviews","availability_365"], hue="room_type")
```

### Geographical distribution of AirBnb Listing

```python
sns.scatterplot(data = df2, x="longitude", y="latitude", hue="room_type")
plt.title("Geographical distribution of AirBnb Listing")
```

### Correlation between different numerical values

```python
corr = df2[["latitude","longitude","price","minimum_nights","number_of_reviews","reviews_per_month","availability_365","beds"]].corr()
corr
```

### HeatMap on Corr

```python
plt.figure(figsize=(16,8))
sns.heatmap(data=corr, annot=True)
```
